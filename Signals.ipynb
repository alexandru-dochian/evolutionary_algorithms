# Example of signals used in Mutation and Crossover

## Example of mutation signal
```
Higher frequency band, corresponding to human brain's Beta waves [12Hz - 35Hz]
The generated signal will be summed over individual's numeric representation
```

````python
from matplotlib import pyplot
from evoman_wrapper.utils.ComputationUtils import ComputationUtils

mutation_signal_config = {
        "number_of_sine_functions": 10,
        "min_frequency": 12,
        "max_frequency": 35,
        "min_amplitude": -0.1,
        "amplitude_range": 0.2
}

signal = ComputationUtils.generate_signal(mutation_signal_config, 100)
pyplot.plot(signal)
````

## Example of crossover signal
```
Lower frequency
Values centered in 0
Acts like a multiple point crossover
```

````python
from matplotlib import pyplot
from evoman_wrapper.utils.ComputationUtils import ComputationUtils

crossover_signal_config = {
        "number_of_sine_functions": 3,
        "min_frequency": 0,
        "max_frequency": 10,
        "min_amplitude": -1,
        "amplitude_range": 2
}

signal = ComputationUtils.generate_signal(crossover_signal_config, 100)
pyplot.plot(signal)
````
