# Example of signals used in Mutation and Crossover

## Example of mutation signal
```
About brain waves:
    - Usual amplitudes of brain waves range from 0 to 100 microVolts.
        {
            "min_amplitude": 0,
            "amplitude_range": 100    
        }
        
        
    - Usual braiwaves frequencies range from roughly 0Hz in `Low delta` to roughly 100Hz in `High Gamma`  
        {
            "min_frequency": 0,
            "max_frequency": 100,
        }
```

````python
from matplotlib import pyplot
from evoman_wrapper.utils.ComputationUtils import ComputationUtils

mutation_signal_config = {
        "number_of_sine_functions": 4,
        "min_frequency": 50,
        "max_frequency": 55,
        "min_amplitude": 0,
        "amplitude_range": 100
}

signal = ComputationUtils.generate_signal(mutation_signal_config, 5000)
pyplot.plot(signal)
````

## Example of crossover signal
```
Lower frequency
Values centered in 0
Acts like a multiple point crossover
```

````python
from matplotlib import pyplot
from evoman_wrapper.utils.ComputationUtils import ComputationUtils

crossover_signal_config = {
        "number_of_sine_functions": 4,
        "min_frequency": 0,
        "max_frequency": 10,
        "min_amplitude": -1,
        "amplitude_range": 2
}

signal = ComputationUtils.generate_signal(crossover_signal_config, 5000)
pyplot.plot(signal)
````
